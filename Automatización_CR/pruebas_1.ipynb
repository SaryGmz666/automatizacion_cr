```python
MESES = {
    "ene": "01",
    "feb": "02",
    "mar": "03",
    "abr": "04",
    "may": "05",
    "jun": "06",
    "jul": "07",
    "ago": "08",
    "sep": "09",
    "oct": "10",
    "nov": "11",
    "dic": "12",
}
```

```python
def normalizar_fecha(fecha_str: str) -> str:
    mes, anio = fecha_str.lower().split("-")
    anio = "20" + anio
    return anio + MESES[mes]
```
