```python
import re

def pattern_decryptor(pattern:str)-> list:
    REF_PATTERN = re.compile(
        r"ref\(\s*(\d+)\s*,\s*([a-zA-Z_][a-zA-Z0-9_]*)\s*\)",
        re.IGNORECASE
    )
    return REF_PATTERN.findall(pattern)
```

```python
import re

wb = load_workbook(PATH + name_tamplate + ".xlsx", data_only=False)

REF_REGEX = re.compile(r"ref\(", re.IGNORECASE)

target_cells = []
keys_cells = []

for ws in wb.worksheets:
    for row in ws.iter_rows():
        for cell in row:
            if isinstance(cell.value, str) and REF_REGEX.search(cell.value):
                target_cells.append(cell)
                keys_cells += (pattern_decryptor(cell.value))
                
```
